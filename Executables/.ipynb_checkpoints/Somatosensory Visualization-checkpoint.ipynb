# Analyzing Diva Vocal Tract Somatosensory System

```python
%matplotlib inline
#Imports
from divapy import Diva
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def vt_som_plot(sys,m):
    arts = np.array([m]*13)
    fig, ax = plt.subplots(1,2)
    a,som,c,d = sys.get_audsom(arts)
    ax[0] = sys.plot_outline(arts, axes=ax[0])
    plt.sca(ax[1])
    plt.plot(som)
```

```python
interact(vt_som_plot,
         sys = fixed(Diva()),
         m = widgets.FloatSlider(value=0,
                                 min=-3.,
                                 max=3.,
                                 step=1,
                                 description='Test:',
                                 disabled=False,
                                 continuous_update=False,
                                 orientation='horizontal',
                                 readout=True,
                                 readout_format='i',
                                 slider_color='white'
                                 ))
```
